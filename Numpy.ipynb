##  統計量を求める

```python
std_norm = np.random.randn(5,5)
std_norm
```

```python
#最大値
std_norm.max()
```

```python
np.max(std_norm)
```

```python
#最小値
std_norm.min()
```

```python
np.min(std_norm)
```

```python
#最大値のindexを取得→1行にした時のindex
std_norm.argmax()
```

```python
std_norm.max()
```

```python
std_norm.flatten()[8]
```

```python
#最小値のindexを取得→1行にした時のindex
std_norm.argmin()
```

```python
std_norm.min()
```

```python
std_norm.flatten()[22]
```

```python
#平均値を取得
std_norm.mean()
```

```python
#中央値を取得  max,min,meanと異なり、ndarrayから直接取得できないことに注意
np.median(std_norm)
```

```python
#中央値 vs 平均値
#time.time()で現時点での時間を測定
import time
a = np.random.randn(1000,1000)
before = time.time()
np.median(a)
after_median = time.time()
np.mean(a)
after_mean = time.time()
print('median took {} sec'.format(after_median - before))
print('mean took {} sec'.format(after_mean - after_median))
#中央値は並び替えのせいで平均値より処理に時間がかかる
```

```python
#標準偏差(分散の平方根)を取得
np.std(std_norm)
```

```python
#max,min,meanなどにaxisを指定することで、行毎、列毎の統計量を測定可能→axis=0は列毎、axis=1は行毎
std_norm.max(axis=0)
```
